```python
# Create the arrays for the independent and dependent variables
xmin = 0
xmax = 5
Nx = 11  # Grid size (number of grid points) of the x-axis
x = np.linspace(xmin,xmax,Nx)
```

```python
# Lets define the array for the independent variable
# Function : y =  x^2
y = x**2    # numpy applies operations element-wise
```

```python
# In matplotlib there are two ways to plot:
# Functional method
# Object orientated method (OO) [preferred method]
```

```python
# Functional method
plt.plot(x,y)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Title')
```

```python
# Plotting the function y = x^3 + 1

# Creating the arrays for the dependent and independent variables
xmin = 0
xmax = 10
Nx = 12
x = np.linspace(xmin,xmax,Nx)
```

```python
x
```

```python
len(x)
```

```python
# Defining the array for the independent variables
# Function is y = x^3 + 1  # numpy applies operation element-wise
y = x**3 + 1
```

```python
y
```

```python
# Plotting using the Funtional method
plt.plot(x,y)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Title')
```

```python
# Object-Orientated (OO) Method
# Create a figure object: "Blank object"
# Run method/operations using dot notation
lower_left_x = 0
lower_left_y = 0
width = 1
height = 1
fig = plt.figure()   # This command creates the object. Instatiate
axes = fig.add_axes([lower_left_x,lower_left_y,width,height])
axes.plot(x,y)
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('My first OO plot')
```

```python
# OO method
# Create a fig method: 'blank object'
# Create a plot which is excatly as big as the canvas
lower_left_x = 0
lower_right_y = 0
width = 1
height = 1
fig = plt.figure() # This is the figure plot


# Set up origin and size of axes 1
axes1 = fig.add_axes([lower_left_x,lower_right_y,width,height])

# set up the origin and sizes for axes 2 (small plot)
perct_in = 0.2  # Percentage inward from axes 1
perct_up = 0.5  # Percentage upward from axes 1
perct_width = 0.3
perct_height = 0.3
axes2 = fig.add_axes([perct_in,perct_up,perct_width,perct_height])

# Plot on axes 1
axes1.plot(x,y,'b')
axes1.set_xlabel('x label 1')
axes1.set_ylabel('ylabel 1')
axes1.set_title('Title label 1')

# Plot on axes 2
axes2.plot(y,x,'r')
axes2.set_xlabel('x label 2')
axes2.set_ylabel('y label 2')
axes2.set_title('Title label 2')
```

# Creating subplots with the OO method


```python
# As the number of rows and columns increase, the figures overlapp
# This is fixed with 'tight_layout command'
fig, axes = plt.subplots(nrows=2,ncols=4)
```

```python
# Creating subplots with OO method
fig, axes = plt.subplots(nrows=2,ncols=2)
plt.tight_layout()
```

```python
axes
```

```python
# Subplot with OO method
# This example shows the power of subplots
# We can loop over every axes with a 'for' loop
fig, axes = plt.subplots(nrows=1,ncols=2)
plt.tight_layout()

for current_ax in axes:
    current_ax.plot(x,y,'b')
```

```python
# OO method for subplots
# Not going to use 'for' loo
# Explicitly label the array with an index
fig, axes = plt.subplots(nrows=1,ncols=2)

# Plot on the 1st element (index=0) of axes array
axes[0].plot(x,y,'b')
axes[0].set_xlabel('x label: index 0')
axes[0].set_ylabel('y label: index 0')
axes[0].set_title('title: index 0')

axes[1].plot(y,x,'r')
axes[1].set_xlabel('x label: index 0')
axes[1].set_ylabel('y label: index 0')
axes[1].set_title('title: index 0')

plt.tight_layout()
```

```python
# OO method for subplots
# Specify the width and height
width = 5
height = 5
number_rows  = 2
number_columns = 1
fig, axes = plt.subplots(nrows=number_rows,ncols=number_columns, figsize=(width,height))

print('axes object: ',axes)
print('length of axes object: ', len(axes))

axes[0].plot(x,y,'b')
axes[0].set_xlabel('x label: index 0')
axes[0].set_ylabel('y label: index 0')
axes[0].set_title('Title: index 0')

axes[1].plot(y,x,'r')
axes[1].set_xlabel('x label: index 0')
axes[1].set_ylabel('y label: index 0')
axes[1].set_title('Title: index 0')
plt.tight_layout()
```

```python
# OO method
# Add a legend

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,'b',label='x^2')
ax.plot(x,x**3,'r',label='x^3')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Title')
ax.legend(loc=0)
# ax.legend(loc=(0.1,0.1)) specify postion of legend
```
