```python
## C is the ticker for CityGroup
C = data.DataReader("C",'yahoo', start, end)

## GS is the ticker for Goldman Sachs
GS = data.DataReader("GS",'yahoo', start, end)

## JPM is the ticker for JPMorgan Chase
JPM = data.DataReader("JPM",'yahoo', start, end)

## MS is the ticker for Morgan Stanley
MS = data.DataReader("MS",'yahoo', start, end)

## WFC is the ticker for Wells Fargo.
WFC = data.DataReader("WFC",'yahoo', start, end)
```

Create a list of the ticker symbols (as strings) in alphabetical order. Call this list as: tickers

```python
tickers = ['BAC','C','GS','JPM','MS','WFC']
```

Use pd.concat to concatenate the bank dataframes together into a single dataframe called bank_stocks. Set the keys argument equal to the tickers list. Also pay attention to the axis you concatenate.

```python
bank_stocks = pd.concat([BAC,C,GS,JPM,MS,WFC], axis=1 ,keys=tickers)
```

```python
bank_stocks.head()
```

Set the Column name levels 

```python
bank_stocks.columns.names = ['Bank Ticker', 'Stock Info']
```

```python
bank_stocks.head()
```

# EDA
Exploring the data a bit. 

What is the max close price for each bank's stock throughout the time period?

```python
for tick in tickers:
    print(tick,bank_stocks[tick]['Close'].max())
```

```python
## Another savy way of doing this is using the syntax (.xs)
bank_stocks.xs(key='Close', axis=1, level='Stock Info').max()
```

Create a new DataFrame called returns. This dataframe will contain the returns for each bank's stocks.

```python
returns = pd.DataFrame()
```

We can use pandas pct_change() method on the close column to create a column representing this return value. Create a for loop that goes and for each Bank Stock Ticker creates this return column and sets it as a column in the return DataFrame.


```python
for tick in tickers:
    returns[tick+'Return'] = bank_stocks[tick]['Close'].pct_change()
```

```python
returns.head()
```

Create a pairplot using seaborn of the return dataframe. What stock stands out to you? Can you find out why?

```python
import seaborn as sns
sns.pairplot(returns[1:])
```

# Moving Averages

Let us analyze the moving averages for stocks in the year 2008

# Plot the rolling 30 day average against the Close Price for Bank of America's stock for the year 200  

```python
plt.figure(figsize=(15,5))
BAC['Close'].loc['2008-01-01':'2009-01-01'].rolling(window=30).mean().plot(label='30 day Moving Average')
BAC['Close'].loc['2008-01-01':'2009-01-01'].plot(label='BAC Close')
plt.legend()
```

Create a heatmap of the correlation between the stocks Close Price.

```python
sns.heatmap(bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr(), annot=True)
```

Optional use seaborn's clustermap to cluster the correlations together.
